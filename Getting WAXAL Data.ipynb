```python
from huggingface_hub import login, HfApi

# ── Authenticate ──────────────────────────────────────────
TOKEN_PATH = r"C:\Users\Chidiebere\Documents\IgboTTS\huggingfaceAPI.txt"

with open(TOKEN_PATH, "r") as f:
    hf_token = f.read().strip()

login(hf_token)

# ── List all files in the WAXAL repo directly ─────────────
# This bypasses get_dataset_config_names completely
api = HfApi()
repo_files = api.list_repo_files("google/WaxalNLP", repo_type="dataset")

# Filter for Igbo-related files
igbo_files = [f for f in repo_files if "ibo" in f.lower()]

print("All repo files:")
for f in sorted(repo_files):
    print(f"  {f}")

print(f"\nIgbo-related files found:")
for f in igbo_files:
    print(f"  {f}")
```

```python
from datasets import load_dataset

# ── Load WAXAL Igbo TTS using data_dir (not a config name) ──
ds = load_dataset(
    "google/WaxalNLP",
    data_dir="data/TTS/ibo",       # exact folder path from the repo
    token=hf_token,                # your already-loaded token
)

print(ds)

# ── Inspect a sample ─────────────────────────────────────────
sample = ds["train"][0]
print("\nFields available:", list(sample.keys()))
print("Text:", sample["text"])
print("Audio sampling rate:", sample["audio"]["sampling_rate"])
print("Audio array length:", len(sample["audio"]["array"]))
```

```python
import pandas as pd
from huggingface_hub import hf_hub_download

# ── Download each parquet file directly ──────────────────
files = [
    "data/TTS/ibo/ibo-train-00000.parquet",
    "data/TTS/ibo/ibo-train-00001.parquet",
    "data/TTS/ibo/ibo-validation-00000.parquet",
    "data/TTS/ibo/ibo-test-00000.parquet",
]

dfs = []
for f in files:
    path = hf_hub_download(
        repo_id="google/WaxalNLP",
        filename=f,
        repo_type="dataset",
        token=hf_token,
    )
    dfs.append(pd.read_parquet(path))
    print(f"Loaded: {f}")

# ── Combine all splits ────────────────────────────────────
df = pd.concat(dfs, ignore_index=True)

print(f"\nTotal records : {len(df)}")
print(f"Columns       : {list(df.columns)}")
print(f"\nSample row:")
print(df.iloc[0])
```
